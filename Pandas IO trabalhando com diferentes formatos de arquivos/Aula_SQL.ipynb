# Criando nosso banco de dados

Agora, precisamos fazer algumas importações. Na linha de baixo, escrevemos from sqlalchemy import. A primeira importação será create_engine, que serve para criar o motor, ou seja, a conexão com o banco de dados que o fará funcionar.

Na mesma linha, inserimos uma vírgula e a importação MetaData. Essa é uma classe que representa os meta dados do banco de dados, isso inclui informações sobre as tabelas, colunas e outros objetos.

Também importamos a classe Table que representa uma tabela em um banco de dados e é usada para manipular dados em uma tabela específica.

Por fim, importamos a função inspect, que permite inspecionar a estrutura do banco de dados. Em seguida, executamos a célula.

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Agora, vamos criar o banco de dados. Para isso, criamos uma variável chamada engine. Dela, chamamos o create_engine(). Dentro das chaves abrimos aspas simples e escrevemos o banco de dados que será utilizado, 'sqlite:///:memory:'.

Nesse caso, usaremos o SQLite, um banco de dados que vem naturalmente instalado no Google Colab. Nesse trecho de código o memory indica que estamos usando um banco de dados local.

```python
engine = create_engine('sqlite:///:memory:')
```

# Obtendo os Dados

```python
url = ('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

# Salvando no Banco de Dados

```python
dados.to_sql('Clientes', engine, index=False)
```

# Inspecinando o Banco de Dasdos

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

## As instruções SQL podem ser compostas por uma ou mais cláusulas que fornecem informações adicionais sobre o que a consulta deve fazer. As cláusulas mais comuns são:

SELECT: especifica quais colunas devem ser selecionadas na consulta.

FROM: especifica as tabelas do banco de dados que devem ser consultadas.

WHERE: filtra os resultados da consulta com base em uma ou mais condições especificadas.

ORDER BY: classifica os resultados da consulta em ordem crescente ou decrescente com base em uma ou mais colunas.

GROUP BY: agrupa os resultados da consulta com base em uma ou mais colunas.

LIMIT: limita o número de linhas retornadas pelos resultados da consulta.

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

## Salvando a nova tabela com filtro

```python
empregados.to_sql('empregados', con=engine, index=False)
```

## Leitura de uma tabela inteira

```python
pd.read_sql_table('empregados', engine)
```

## Ler uma tabela  (colunas)

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

# Atualizando um banco de dados

##  Deletando um dado

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('Clientes', engine)
```

## Atualizando Dados

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```
